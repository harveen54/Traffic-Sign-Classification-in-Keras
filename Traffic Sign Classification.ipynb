```python
X_images =[]
y_labels =[]

for image in list_images:
    #print(image)
    
    #for input data
    local_image = cv2.cvtColor( cv2.imread(image) , cv2.COLOR_BGR2RGB)
    X_images.append(local_image)
    
    #for labels
    class_traffic_sign = image.split('/')[2]
    class_traffic_sign_slice = class_traffic_sign[len(class_traffic_sign)-2:len(class_traffic_sign)]
    class_traffic_sign_int = int(class_traffic_sign_slice)
    y_labels.append(class_traffic_sign_int)
```

```python
print(len(X_images) , len(y_labels))
```

```python
print(X_images[0])
print(np.array(X_images[0]))
```

```python
X = np.array(X_images)
y = np.array(y_labels)
```

```python
print(X.shape , y.shape)
```

```python
X_resized = []
for image in X:
    resized_image = cv2.resize(image, (64,64))
    X_resized.append(resized_image)
```

```python
X_resized = np.array(X_resized)
X_resized.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_resized , y, test_size=0.1 , shuffle=True)
```

```python
print('X in Training set ' ,X_train.shape)
print('y in Training set ', y_train.shape)
print('X in Test set ', X_valid.shape)
print('y in Test set' , y_valid.shape)
```

```python
np.min(y_train)
```

```python
import keras
from keras.layers.core import Flatten,Dense,Dropout, Activation, Lambda
from keras.models import Model, Sequential
from keras.optimizers import Adam,SGD
from keras.callbacks import TensorBoard
```

```python
tboard = TensorBoard(log_dir = './logs/')
```

```python
model = Sequential()
model.add(Lambda(lambda x: (x / 255.0) - 0.5, input_shape = (64,64,3)))
#model.add(Lambda(lambda x: x+0.1 , input_shape = (64,64,3)))
model.add(Flatten())
model.add(Dense(10))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(100))
model.add(Activation('relu'))
#model.add(Dropout(0.2))
model.add(Dense(62))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer = 'Adam' , loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30, verbose =1, validation_data=(X_valid, y_valid) , batch_size = 64, callbacks=[tboard])
```
